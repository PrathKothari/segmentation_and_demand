```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import numpy as np

def create_sequences(features, target, time_step=1):
    X, y = [], []
    for i in range(len(features) - time_step - 1):
        X.append(features[i:(i + time_step), :])
        y.append(target[i + time_step])
    return np.array(X), np.array(y)

def train_lstm_model(features, target, time_step=10, batch_size=32, epochs=10):
    # Separate scalers for features and target
    feature_scaler = MinMaxScaler(feature_range=(0, 1))
    target_scaler = MinMaxScaler(feature_range=(0, 1))

    scaled_features = feature_scaler.fit_transform(features)
    scaled_target = target_scaler.fit_transform(target.values.reshape(-1, 1))
    
    X, y = create_sequences(scaled_features, scaled_target, time_step)
    
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    model = Sequential()
    model.add(LSTM(100, return_sequences=True, input_shape=(time_step, X.shape[2])))
    model.add(Dropout(0.2))
    model.add(LSTM(100, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))  # output layer

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.summary()

    # Callbacks for early stopping and learning rate reduction
    early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=0.001)

    # Train the model
    history = model.fit(X_train, y_train, 
                        epochs=epochs, 
                        batch_size=batch_size, 
                        validation_data=(X_test, y_test), 
                        verbose=1,
                        callbacks=[early_stop, lr_scheduler])

    return model, history, X_test, y_test, target_scaler

def evaluate_lstm_model(model, X_test, y_test, scaler):
    y_pred_scaled = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred_scaled)
    y_test_orig = scaler.inverse_transform(y_test)
    
    mse = mean_squared_error(y_test_orig, y_pred)
    mae = mean_absolute_error(y_test_orig, y_pred)
    r2 = r2_score(y_test_orig, y_pred)
    rmse = np.sqrt(mse)
    
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"Root Mean Squared Error (RMSE): {rmse}")
    print(f"R-Squared (R2): {r2}")
    
    return y_pred, mse, mae, rmse, r2

# Sample run
train_size = 0.8
train_samples = int(train_size * len(merged_data))
train_data = merged_data[:train_samples]
val_data = merged_data[train_samples:]

features = merged_data.drop(columns=['Weekly_Sales'])
target = merged_data['Weekly_Sales']

model, history, X_test, y_test, scaler = train_lstm_model(features, target, epochs=10)

y_pred, mse, mae, rmse, r2 = evaluate_lstm_model(model, X_test, y_test, scaler)
```

```python

import numpy as np

def forecast_with_lstm_model(model, last_known_data, n_future_steps, scaler, time_step=10):
    """
    Forecast future values using a trained LSTM model.

    Parameters:
    - model: Trained LSTM model
    - last_known_data: Last available data (features) in the shape (1, time_step, n_features)
    - n_future_steps: Number of future steps to forecast
    - scaler: Scaler used to scale the data (both features and target)
    - time_step: Number of time steps used for sequences

    Returns:
    - future_predictions: Array of forecasted values
    """
    future_predictions = []

    current_input = last_known_data

    for _ in range(n_future_steps):
        # Predict the next step
        next_pred_scaled = model.predict(current_input)
        
        # Reshape prediction to match feature dimensions (1, n_features)
        next_pred_scaled = next_pred_scaled.reshape(-1, 1)  # Ensure shape (batch_size, 1)
        
        # Inverse transform the prediction to original scale
        next_pred = scaler.inverse_transform(next_pred_scaled)
        
        # Append the prediction to future predictions
        future_predictions.append(next_pred[0, 0])

        # Update current input: remove the first step and add the predicted step
        current_input_reshaped = current_input[0]  # Shape (time_step, n_features)
        
        # Create a new array for the next input with correct dimensions
        next_pred_expanded = np.zeros((1, current_input_reshaped.shape[1]))  # Shape (1, n_features)
        next_pred_expanded[0, 0] = next_pred[0, 0]  # Place the predicted value in the correct position
        
        # Concatenate to create new input sequence
        next_input = np.vstack([current_input_reshaped[1:], next_pred_expanded])  # Shape (time_step, n_features)
        
        # Add back batch dimension
        current_input = next_input.reshape(1, time_step, next_input.shape[1])  # Shape (1, time_step, n_features)

    return np.array(future_predictions)

# Example usage:
# last_known_data shape should be (1, 10, 25)
# Number of future steps you want to predict
```
