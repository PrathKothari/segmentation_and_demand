```python
#Converting Day to datetime
start_date = pd.Timestamp('2021-01-01')
final_df['DATE'] = start_date + pd.to_timedelta(final_df['DAY']-1, unit='D')
```

```python
current_date = final_df['DATE'].max()
```

```python
final_df.drop(columns=['DAY'], inplace=True)
```

```python
rfm_df = final_df.groupby('household_key').agg({
    'DATE': lambda x: (current_date - x.max()).days, # Recency
    'BASKET_ID': 'count', #Frequency
    'SALES_VALUE': 'sum'

}).reset_index()
rfm_df.columns = ['household_key', 'Recency', 'Frequency', 'Monetary']
rfm_df.head()
```

```python
rfm_df['Recency_Score'] = pd.qcut(rfm_df['Recency'], 4, labels=False) + 1
rfm_df['Frequency_Score'] = pd.qcut(rfm_df['Frequency'], 4, labels=False) + 1
rfm_df['Monetary_Score'] = pd.qcut(rfm_df['Monetary'], 4, labels=False) + 1
```

```python
rfm_df['RFM_Score'] = rfm_df['Recency_Score'].astype(str) + rfm_df['Frequency_Score'].astype(str) + rfm_df['Monetary_Score'].astype(str)
```

```python
rfm_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming RFM scores are on different scales, normalize them
scaler = MinMaxScaler()
rfm_df[['Recency_Score', 'Frequency_Score', 'Monetary_Score']] = scaler.fit_transform(
    rfm_df[['Recency_Score', 'Frequency_Score', 'Monetary_Score']]
)

# Sum the normalized scores to create a composite score
rfm_df['RFM_Score'] = rfm_df['Recency_Score'] + rfm_df['Frequency_Score'] + rfm_df['Monetary_Score']
#Normalization
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(rfm_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

This indicates that we have to take 4 clusters as theres a sharp decrease between 2 and 4.

```python
# Run KMeans with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_df[['Recency_Score', 'Frequency_Score', 'Monetary_Score']])

# Now you can analyze the clusters
# print(rfm_df.groupby('Cluster').mean())
print(rfm_df.groupby('Cluster').mean())
```

```python
# Calculate percentage changes
final_df['Price_Change'] = final_df.groupby('PRODUCT_ID')['SALES_VALUE'].pct_change()
final_df['Quantity_Change'] = final_df.groupby('PRODUCT_ID')['QUANTITY'].pct_change()

# Filter out rows where Price_Change is zero or close to zero to avoid division by zero
final_df = final_df[(final_df['Price_Change'] != 0) & (final_df['Price_Change'].abs() > 1e-6)]

# Calculate elasticity
final_df['Price_Elasticity'] = final_df['Quantity_Change'] / final_df['Price_Change']

# Handle NaN and inf values: drop them or fill with a default value (e.g., 0 or mean elasticity)
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
final_df.dropna(subset=['Price_Elasticity'], inplace=True)

# Analyze elasticity by cluster
elasticity_by_cluster = final_df.groupby('Cluster_Label')['Price_Elasticity'].mean()
print(elasticity_by_cluster)
```

Insights:

The elasticity values are positive and relatively similar across all clusters. This suggests that, on average, a price increase is associated with a proportional decrease in quantity purchased, but the sensitivity is not extreme.
A price elasticity close to 1 would suggest that customers are somewhat sensitive to price changes. However, the values being slightly less than 1 indicate that the customers in these segments are relatively price inelastic, meaning price changes do not drastically affect their purchasing behavior.
Segment Comparisons:

Old Low-Value Customers have the highest elasticity (0.674618), indicating they might be slightly more sensitive to price changes than other segments.
Recent High-Value Customers have the lowest elasticity (0.653860), suggesting they might be the least price-sensitive segment.
